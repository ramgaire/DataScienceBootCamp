```python
# # this is the case study for managing employess
# # # name,salary,id, age, job position
# # # Employee master
# # class, and function insid ethe calss
# # user can add employee
# # deletes employess
# # update name
# # update salary
# # # update by name
# # update by id
# get details by id
# # det details by name
# one class
# create acalss and simply add the employee
```

```python
class Employee:
  def __init__(self, emp_id, name, position, salary,age):
    self.emp_id = emp_id
    self.name = name
    self.position = position
    self.salary = salary
    self.age = age

  def display_info(self):
    print("Employee ID", self.emp_id)
    print("Name", self.name)
    print("Position", self.position)
    print("Salary", self.salary)
    print("Age", self.age)

class EmployeeManagementSystem:
  def __init__(self):
    self.employees = []

  def add_employee(self, employee):
    self.employees.append(employee)

  def remove_employee(self, emp_id):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        self.employees.remove(employee)
        return True
    return False

  def update_name(self,emp_id,name):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.name= name
        return True
    return False

  def update_salary(self,emp_id,salary):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.salary= salary
        return True
    return False

  def update_position(self,emp_id,position):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.position= position
        return True
    return False

  def update_age(self,emp_id,age):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.age= age
        return True
    return False

  def delete_by_name(self,name):
    all_matching_names=[]
    for employee in self.employees:
      if employee.name == name:
        all_matching_names.append(employee)
    if len(all_matching_names)==0:
      return False
    if len(all_matching_names)==1:
      self.employees.remove(all_matching_names[0])
      return True
    if len(all_matching_names)>1:
      print("Duplicate names found please select the id you want to remove")
      all_ids=[]
      for emps in all_matching_names:
        print("ID :- ",emps.emp_id)
        all_ids.append(emps.emp_id)
      id_to_remove=int(input("Please choose the id you want to remove :- "))
      if id_to_remove in all_ids:
        self.employees.remove(all_matching_names[all_ids.index(id_to_remove)])
        return True
      else:
        return False

  def details_by_id(self,emp_id):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.display_info()

  def details_by_name(self,name):
    for employee in self.employees:
      if employee.name == name:
        employee.display_info()
  def details_all_employees(self):
    print("Check length",len(self.employees))
    for employee in self.employees:
        employee.display_info()
        print("====================")
```

```python
controller_object = EmployeeManagementSystem()
while True:
  print("Please choose the option")
  print("1) Add Employee")
  print("2) Delete Employee by ID")
  print("3) Delete Employee by Name")
  print("4) update name by id")
  print("5) update age by id")
  print("6) update salaray by id ")
  print("7) updates position by id")
  print("8) gives details by name")
  print("9) All details")
  print("10) exit the program")
  choice=input("Please choose an option ")
  if choice=='1':
    name=choice=input("Please enter Employee name ")
    salary=choice=input("Please enter Employee salary ")
    age=choice=input("Please enter Employee age ")
    position=choice=input("Please enter Employee position ")
    id=len(controller_object.employees)
    emp_temp=Employee(id,name,position,salary,age)
    controller_object.add_employee(emp_temp)
  if choice=='2':
    id=int(input("Please enter Employee id you want to remove "))
    if controller_object.remove_employee(id):
      print("Employee Removed Successfully")
    else:
      print("Id not found could not remove the employee")
  if choice =='3':
    name = input("please enter the name of employee you want to delete")
    if controller_object.delete_by_name(name):
      print("Employee Removed Successfully")
    else:
      print("Name not found could not remove the employee")

  if choice =='4':
    id =int(input(" enter the  id of customer"))
    name = input("enter the nam eof customer")
    if controller_object.update_name(id,name):
      print("Employee update Successfully")
    else:
      print("not update employee")
  if choice =='5':
    id = int(input(" enter the  id of customer"))
    age = input("enter the age of customer")
    if controller_object.update_age(id,age):
      print("Employee update age by id  Successfully")
    else:
      print("not update age by id employee")

  if choice =='6':
    id = int(input("ente the id of customr"))
    salary = input("enter the salary ")
    if controller_object.update_salary(id,salary):
      print("Employee update salary by id  Successfully")
    else:
      print("not update age by id employee")

  if choice =='7':
    id = int(input("ente the id of customr"))
    position = input("enter the position ")
    if controller_object.update_position(id,position):
      print("Employee update position by id  Successfully")
    else:
      print("not update position by id employee")

  if choice =='8':
    name = input("enter the name of customer")
    if controller_object.details_by_name(name):
      print("Employee info by name")

  if choice=='9':
    controller_object.details_all_employees()

  elif choice =='10':
    break
```
